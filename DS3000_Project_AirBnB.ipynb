```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split

# Load data
df_airbnb_2019 = pd.read_csv('/mnt/data/2019AirBnB.csv')
df_airbnb_2022 = pd.read_csv('/mnt/data/2022Listings.csv')

# Function to preprocess a dataset
def preprocess_data(df):
    # Handling missing values for numeric features
    numeric_features = df.select_dtypes(include=['int64', 'float64']).columns
    for feature in numeric_features:
        df[feature].fillna(df[feature].mean(), inplace=True)

    # Handling categorical features with One-Hot Encoding
    categorical_features = df.select_dtypes(include=['object']).columns
    df = pd.get_dummies(df, columns=categorical_features, drop_first=True)

    # Feature scaling
    scaler = StandardScaler()
    scaled_df = scaler.fit_transform(df)

    return pd.DataFrame(scaled_df, columns=df.columns)

# Preprocess the 2019 and 2022 datasets
processed_2019 = preprocess_data(df_airbnb_2019)
processed_2022 = preprocess_data(df_airbnb_2022)

# Split data into features (X) and target (y) - assuming 'price' is the target variable
# For 2019 dataset
X_2019 = processed_2019.drop('price', axis=1)
y_2019 = processed_2019['price']

# Split 2019 data into training and testing sets
X_train_2019, X_test_2019, y_train_2019, y_test_2019 = train_test_split(X_2019, y_2019, test_size=0.2, random_state=42)

# For 2022 dataset
X_2022 = processed_2022.drop('price', axis=1)
y_2022 = processed_2022['price']

# Split 2022 data into training and testing sets
X_train_2022, X_test_2022, y_train_2022, y_test_2022 = train_test_split(X_2022, y_2022, test_size=0.2, random_state=42)
```
